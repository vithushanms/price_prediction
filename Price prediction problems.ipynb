```python
import numpy as np
```

```python
from sklearn.datasets import load_boston

#load the dataset into x and y axis variables 
load_boston = load_boston();
x = load_boston.data
y = load_boston.target

print(load_boston.DESCR) #print description of the data
```

```python
import pandas as pd

#create a data frame
data = pd.DataFrame(x, columns=load_boston.feature_names); # load all the feature data into the dataframe
data["Market_Price"] = y #create a new column for the outcome/target value within the data frame 
```

```python
#Build model

from sklearn.model_selection import train_test_split

x = data.drop("Market_Price", axis=1)
y = data["Market_Price"]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42) 
```

```python
from sklearn.linear_model import LinearRegression
lr  = LinearRegression()
lr.fit(x_train,y_train)
```

```python
predictions = lr.predict(x_test)

print("the actual market price", y_test[0])
print("the predicted market price", predictions[0])
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
print(rmse)
```
